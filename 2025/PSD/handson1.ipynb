```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
f =  5 #hz
t = np.linspace(0, 2, 0.0001)
y = np.sin(2*np.pi*f*t)
y2 = np.cos(2*np.pi*f*t)

plt.plot(t, y, label='sin')
plt.plot(t, y2, label='cos')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Contoh plot sinus dan cosinus')
plt.legend()
plt.show()   

```
