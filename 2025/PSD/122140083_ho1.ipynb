## **1. Membuat Gelombang Sinus dan Cosinus**
- Buat array t yang berkisar dari 0 hingga 2 dengan langkah sebesar 0.0001.
- Buat sinyal-sinyal berikut menggunakan array waktu ini:
$$
\begin{aligned}
A. \quad y_1 &= 2 \cdot \sin(2\pi \cdot 3 \cdot t + 0) \\
B. \quad y_2 &= 1 \cdot \cos(2\pi \cdot 4 \cdot t + \frac{\pi}{4}) \\
C. \quad y_3 &= -1 \cdot \sin(2\pi \cdot 5 \cdot t + \frac{\pi}{2}) \\
D. \quad y_4 &= 0.5 \cdot \cos(2\pi \cdot 6 \cdot t + \pi) \\
\end{aligned}
$$


```python
# Siapin dulu Librarynya yang akan kita gunakan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Buat Array t yang berkisar dari 0 sampai 2 dengan interval 0.0001
t = np.arange(0, 2, 0.0001)

# Buat sinyal" diatas
y1 = 2 * np.sin(2 * np.pi * 3 * t)
y2 = 1 - np.cos(2 * np.pi * 4 * t + np.pi/4)
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi/2)
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)
```

## **2. Perbandingan Subplot**
- Buat gambar dengan 4 subplot (grid 2x2) untuk membandingkan semua sinyal secara berdampingan.
- Setiap subplot harus berisi salah satu sinyal ( y1 ), ( y2 ), ( y3 ), dan ( y4 ).

```python
# Membuat gambar dengan 4 subplot (grid 2x2) untuk membandingkan semua sinyal secara berdampingan
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.suptitle('Perbandingan Sinyal')
# Sinyal y1
axs[0, 0].plot(t, y1, label='y1 = 2*sin(23t)')
axs[0, 0].axhline(0, color='gray', linestyle='--', linewidth=1)
axs[0, 0].set_title("Sinyal y1")
axs[0, 0].legend(loc='upper right')

# Sinyal y2
axs[0, 1].plot(t, y2, label='y2 = 1 - cos(24t + /4)', color='r')
axs[0, 1].axhline(1, color='gray', linestyle='--', linewidth=1)
axs[0, 1].set_title("Sinyal y2")
axs[0, 1].legend(loc='upper right')

# Sinyal y3
axs[1, 0].plot(t, y3, label='y3 = -sin(25t + /2)', color='g')
axs[1, 0].axhline(0, color='gray', linestyle='--', linewidth=1)
axs[1, 0].set_title("Sinyal y3")
axs[1, 0].legend(loc='upper right')

# Sinyal y4
axs[1, 1].plot(t, y4, label='y4 = 0.5*cos(26t + )', color='m')
axs[1, 1].axhline(0, color='gray', linestyle='--', linewidth=1)
axs[1, 1].set_title("Sinyal y4")
axs[1, 1].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
